```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
import plotly.figure_factory as ff
```

Durante a aula nossa variável a ser prevista (churn) continha apenas 2 categorias, a base de vocês contém mais. O Naive Bayes pode ser aplicado para problemas de classificação com múltiplas classes da mesma forma que para problemas de classificação binária. O Naive Bayes é um algoritmo de classificação probabilístico que calcula a probabilidade de uma amostra pertencer a cada classe e seleciona a classe com a maior probabilidade como a previsão final.
Em resumo, o Naive Bayes pode ser aplicado da mesma maneira para problemas de classificação com múltiplas classes, e os mesmos princípios se aplicam em termos de treinamento, avaliação e aplicação do modelo.

Verifique se o número de linhas condiz, se as variáveis estão corretas sendo apenas a de score para y e as demais nas bases de X e por último, se Y está balanceada no teste.

```python
X_test = pd.read_csv("X_test_Credit_Score_Parte2.csv", delimiter=',')
X_train = pd.read_csv("X_train_Credit_Score_Parte2.csv", delimiter=',')
y_test = pd.read_csv("y_test_Credit_Score_Parte2.csv", delimiter=',')
y_train = pd.read_csv("y_train_Credit_Score_Parte2.csv", delimiter=',')
```

# 2) Aplique o algoritmo de Naive Bayes aos dados de treinamento.

```python
naive_churn = GaussianNB()

naive_churn.fit(X_train, y_train)
```

# 3) Faça a avaliação do modelo com os dados de treinamento.
Traga a acurácia, recall e plote a matriz de confusão. Não se esqueça de avaliar com suas palavras o desempenho do modelo, interpretando as métricas.

Dica: Para calcularmos o recall em classificação multi classe precisamos usar o atributo macro:
recall = recall_score(y_train, y_pred_train, average='macro')

```python
y_pred_train = naive_churn.predict(X_train)
accuracy = accuracy_score(y_train, y_pred_train)
recall = recall_score(y_train, y_pred_train, average='macro')
print("Acurácia:", accuracy)
print("Recall:", recall)
```

```python
conf_matrix = confusion_matrix(y_train, y_pred_train)

class_names = ['High Score', 'Avarege Score', 'Low Score']

# Plotando a matriz de confusão com Plotly
fig = ff.create_annotated_heatmap(
    z=conf_matrix,
    x=class_names,
    y=class_names,
    colorscale='Blues',
    showscale=True
)
fig.update_layout(
    title='Matriz de Confusão',
    xaxis_title='Predicted labels',
    yaxis_title='True labels',
    font=dict(
        family='Arial',
        size=12,
        color='black'
    )
)
```

# 4) Aplique o modelo aos dados de teste e realize a avaliação dos resultados, da mesma forma que fez acima. Não se esqueça de avaliar com as suas palavras e comparar o desempenho da base treino com a teste.

```python
naive_churn.fit(X_test, y_test)
```

```python
y_pred_test = naive_churn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test, average='macro')
print("Acurácia:", accuracy)
print("Recall:", recall)
```

```python

conf_matrix = confusion_matrix(y_test, y_pred_test)

class_names = ['High Score', 'Avarege Score', 'Low Score']

# Plotando a matriz de confusão com Plotly
fig = ff.create_annotated_heatmap(
    z=conf_matrix,
    x=class_names,
    y=class_names,
    colorscale='Blues',
    showscale=True
)
fig.update_layout(
    title='Matriz de Confusão',
    xaxis_title='Predicted labels',
    yaxis_title='True labels',
    font=dict(
        family='Arial',
        size=12,
        color='black'
    )
)
```

O modelo apresentou um desempenho excelente na base de treino e um desempenho perfeito na avaliação de métricas da base de teste. Na prática de Machine Learning, é esperado que o desempenho no teste sofra uma leve queda em relação ao treino; um resultado cravado em 100% no teste pode indicar que a base separada para teste é muito pequena, possui padrões idênticos aos do treino

A matriz revela que o modelo aprendeu muito bem os padrões, acertando a esmagadora maioria das classificações na diagonal principal: 88 acertos para "High Score", 89 para "Avarege Score" e 88 para "Low Score". Os erros foram marginais, como, por exemplo, classificar erroneamente 2 casos reais de "Low Score" como "High Score".

Analisando a segunda matriz gerada, o modelo acertou 5 casos reais de "High Score", 23 de "Avarege Score" e 2 de "Low Score". Apesar do bom volume de acertos no "Avarege Score", a matriz revela que o modelo cometeu alguns erros, como prever que 2 casos reais de "Low Score" eram "Avarege Score" e 1 caso real de "Low Score" era "High Score".
