```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
df = pd.read_csv('Credit_Score_Project_Parte2.csv', delimiter=',')
```

```python
df.head()
```

```python
X = df.drop('Credit Score', axis=1)
y = df['Credit Score']
```

```python
X_numerico = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_numerico, y, test_size=0.20, random_state=42)
```

```python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
X_test.to_csv('X_test_Credit_Score_Parte2.csv', index=False)
y_test.to_csv('y_test_Credit_Score_Parte2.csv', index=False)


X_train_smote.to_csv('X_train_Credit_Score_Parte2.csv', index=False)
y_train_smote.to_csv('y_train_Credit_Score_Parte2.csv', index=False)
```
